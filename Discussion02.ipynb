```python
import numpy as np
import pandas as pd
import ipywidgets as widgets

hls_all_raw = pd.read_csv("/content/HSL_12022024195310707.csv")
print(hls_all_raw)
```

```python
print(hls_all_raw["Indicator"])
print("\n===========================================================\n")
hls_slice = pd.DataFrame(hls_all_raw, columns =["Country","Indicator","Type of indicator","Time","Value"])
print(hls_slice)
```

```python
# Filter HLS data for the indicator "Life satisfaction"
hls_ls = hls_slice.loc[hls_all_raw["Indicator"] == "Life satisfaction"]
print(hls_ls)

print("\n===========================================================\n")
print("Total records:")
print(len(hls_ls))

print("\n===========================================================\n")
print("Total Unique Countries:")
print(len(hls_ls["Country"].unique()))

print("\n===========================================================\n")
print("Country List")
print(hls_ls["Country"].unique())
```

```python
# Filtered HLS data for the year 2018 and type of indicator "Average"
hls_train = hls_ls.loc[hls_ls["Time"] == 2018]
hls_train = hls_train.loc[hls_ls["Type of indicator"] == "Average"]

print("\n===========================================================\n")
print("Total records:")
print(len(hls_train))

print("\n===========================================================\n")
print("Total Unique Countries:")
print(len(hls_train["Country"].unique()))

print("\n===========================================================\n")
print("Record:")
print(hls_train)
```

```python
import pandas as pd

# Read the CSV file
weo_raw = pd.read_csv("/content/WEOOct2023all.xls", sep='\t', encoding='latin1')
print(weo_raw)
```

```python
# Drop rows with NaN values in the 'WEO Subject Code' column
weo_raw.dropna(subset=['WEO Subject Code'], inplace=True)


# Select rows where 'WEO Subject Code' contains "BCA_NGDPD"
weo_selected_measurement = weo_raw.loc[weo_raw['WEO Subject Code'].str.contains("BCA_NGDPD")]

# Create a new DataFrame containing only the 'Country' and '2018' columns from the selected rows
weo_selected_measurement_2018 = pd.DataFrame(weo_selected_measurement, columns=['Country', '2018'])

print(weo_selected_measurement_2018)
```

```python
# Merge hls_train with weo_selected_measurement_2018 on the "Country" column
merged_train_data = pd.merge(hls_train, weo_selected_measurement_2018, on="Country")

# Rename columns
merged_train_data = merged_train_data.rename(columns={"Value": "Happiness Measurement", "2018": "Balance of Current Account"})

# Create a new DataFrame containing only the specified columns
merged_train_data = pd.DataFrame(merged_train_data, columns=['Country', 'Happiness Measurement', 'Balance of Current Account'])

print(merged_train_data)
```

```python
import matplotlib.pyplot as plt
import sklearn.linear_model

X = np.c_[merged_train_data["Balance of Current Account"]]
Y = np.c_[merged_train_data["Happiness Measurement"]]
x = X.tolist()
y = Y.tolist()

# plot data
out1 = widgets.Output()
with out1:
  plt.scatter(x, y)
  plt.xlabel('Balance Account')
  plt.ylabel('Happiness')
  plt.title("Data Plot")
  plt.show()

# fit linear model
model = sklearn.linear_model.LinearRegression()
model.fit(X, Y)

# plot predictions
predict_x = [x for x in range(901)]
predict_x = [[x/100] for x in predict_x]
predict_y = model.predict(predict_x)

out2 = widgets.Output()
with out2:
  plt.scatter(predict_x, predict_y)
  plt.scatter(x, y)
  plt.xlabel('Balance Account')
  plt.ylabel('Happiness')
  plt.title("Prediction Line")
  plt.show()

display(widgets.HBox([out1,out2]))
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

y_true = merged_train_data['Happiness Measurement']
y_pred = model.predict(X)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_true, y_pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```
