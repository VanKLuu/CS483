#### Part 1 data preparation

In the cell below, generate training data for your model.
<br> Generate a decent amount of training data in the interval of x.
<br>
<br>All data need to be separated by the same distance.


```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from keras import layers

# Generate data
xc = np.linspace(-100, 100, 1000)
yc = xc * np.sin(xc**2 / 300)

plt.plot(xc, yc)
plt.show()
```

#### Part 2 build models

In the cells below, build your models and train it with data from part1.
<br>You need to split your training data to two parts. With 40% used in training, and 60% used in test.

<br>You may do this with sklearn's train_test_split() or keras' validation_split together with train_test_split().

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Input, Dense
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


x_train, x_test, y_train, y_test = train_test_split(xc, yc, test_size=0.6, random_state=42)

model_single = Sequential([
    Input(shape=(1,)),
    Dense(10, activation='relu'),
    Dense(1)
])

# Compile the model
model_single.compile(loss='mae', optimizer='adam', metrics=['mae'])

# Train the model
history = model_single.fit(x_train, y_train, batch_size=128, epochs=50, validation_split=0.2)

# Predictions on testing data
py = model_single.predict(x_test)

# Plot
plt.scatter(x_test, y_test, label='Actual Data')
plt.scatter(x_test, py, label='Predictions')
plt.legend()
plt.title('Single-layer Network')
plt.show()


```

```python
# model 2: deeper neural network

model_deep = Sequential([
    Input(shape=(1,)),
    Dense(50, activation='relu'),
    Dense(50, activation='relu'),
    Dense(50, activation='relu'),
    Dense(1)
])

# Compile the model
model_deep.compile(loss='mae', optimizer='adam', metrics=['mae'])

# Train the model
history = model_deep.fit(x_train, y_train, batch_size=128, epochs=50, validation_split=0.2)

# Predictions on testing data
py = model_deep.predict(x_test)

# Plot
plt.scatter(x_test, y_test, label='Actual Data')
plt.scatter(x_test, py, label='Predictions')
plt.legend()
plt.title('Deeper-layer Network')
plt.show()
```

```python
# model 3: wider neural network

model_wide = Sequential([
    Input(shape=(1,)),
    Dense(100, activation='relu'),
    Dense(1)
])

# Compile the model
model_wide.compile(loss='mae', optimizer='adam', metrics=['mae'])

# Train the model
history = model_wide.fit(x_train, y_train, batch_size=128, epochs=50, validation_split=0.2)

# Predictions on testing data
py = model_wide.predict(x_test)

# Plot
plt.scatter(x_test, y_test, label='Actual Data')
plt.scatter(x_test, py, label='Predictions')
plt.legend()
plt.title('Wider-layer Network')
plt.show()
```

#### Part 3 model evaluation

In the cells below, eval your model with training data, test data ( with model.evaluate( ) ), and plot your predict result on the same plot with the plot of goal function.

```python
# Evaluate model 1 (single-layer network)
train_loss_1, train_mae_1 = model_single.evaluate(x_train, y_train)
test_loss_1, test_mae_1 = model_single.evaluate(x_test, y_test)

print("Model 1 - Single-layer Network:")
print("Training Loss:", train_loss_1)
print("Training MAE:", train_mae_1)
print("Test Loss:", test_loss_1)
print("Test MAE:", test_mae_1)

# Evaluate model 2 (deeper network)
train_loss_2, train_mae_2 = model_deep.evaluate(x_train, y_train)
test_loss_2, test_mae_2 = model_deep.evaluate(x_test, y_test)

print("\nModel 2 - Deeper-layer Network:")
print("Training Loss:", train_loss_2)
print("Training MAE:", train_mae_2)
print("Test Loss:", test_loss_2)
print("Test MAE:", test_mae_2)

# Evaluate model 3 (wider network)
train_loss_3, train_mae_3 = model_wide.evaluate(x_train, y_train)
test_loss_3, test_mae_3 = model_wide.evaluate(x_test, y_test)

print("\nModel 3 - Wider-layer Network:")
print("Training Loss:", train_loss_3)
print("Training MAE:", train_mae_3)
print("Test Loss:", test_loss_3)
print("Test MAE:", test_mae_3)

# Plotting
plt.figure(figsize=(10, 6))

# Plot original function
plt.plot(xc, yc, label='Goal Function', color='blue')

# Plot predictions from model 1
py_1 = model_single.predict(xc)
plt.scatter(xc, py_1, label='Model 1 Predictions', color='orange')

# Plot predictions from model 2
py_2 = model_deep.predict(xc)
plt.scatter(xc, py_2, label='Model 2 Predictions', color='green')

# Plot predictions from model 3
py_3 = model_wide.predict(xc)
plt.scatter(xc, py_3, label='Model 3 Predictions', color='red')

plt.legend()
plt.title('Comparison of Model Predictions with Goal Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### Part 4 get model output and feedforward by yourself

Recall how a Feedforward Neural Network gets its output. Now choose your model with highest accuracy, and call get_weights( ) to get its weights and bias.
<br> Hint: bias and weights have different dimensions in most cases. Try to guess which index represent bias before you look it up on the internet.

<br> Afterwards, choose 5 data from your training dataset, do all the necessary calculations (with program of course), and get the output of your model. Compare it to the model.predict( ) result.
<br> They should be identical for the first several digitals, if not for all digits.

```python
selected_x = x_train[:5]

# Evaluate models using testing data
test_loss_1, test_mae_1 = model_single.evaluate(x_test, y_test)
test_loss_2, test_mae_2 = model_deep.evaluate(x_test, y_test)
test_loss_3, test_mae_3 = model_wide.evaluate(x_test, y_test)

lowest_test_mae = min(test_mae_1, test_mae_2, test_mae_3)

if lowest_test_mae == test_mae_1:
    best_model = model_single
elif lowest_test_mae == test_mae_2:
    best_model = model_deep
else:
    best_model = model_wide

# Get weights of the best model
weights = best_model.get_weights()
print("Number of weight matrices:", len(weights))

# Perform manual calculations using the obtained weights
manual_outputs = []
for x in selected_x:
    layer_output = np.dot(x, weights[0]) + weights[1]
    layer_output = np.maximum(0, layer_output)
    layer_output = np.dot(layer_output, weights[2]) + weights[3]
    layer_output = np.maximum(0, layer_output)
    output = np.dot(layer_output, weights[-2]) + weights[-1]
    manual_outputs.append(output)

# Predictions using model.predict()
predictions = best_model.predict(selected_x)

# Compare manual calculation with model predictions
for i in range(len(selected_x)):
    print("Data Point:", selected_x[i])
    print("Manual Calculation Output:", manual_outputs[i])
    print("Model Prediction Output:", predictions[i])
    print()

```
