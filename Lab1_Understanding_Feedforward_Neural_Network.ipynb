#### Part 1 data preparation

In the cell below, generate training data for your model.
<br> Generate a decent amount of training data in the interval of x.
<br>
<br>All data need to be separated by the same distance.


```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from keras import layers

# Generate data
xc = np.linspace(-100, 100, 1000)
yc = xc * np.sin(xc**2 / 300)

plt.plot(xc, yc)
plt.show()
```

#### Part 2 build models

In the cells below, build your models and train it with data from part1.
<br>You need to split your training data to two parts. With 40% used in training, and 60% used in test.

<br>You may do this with sklearn's train_test_split() or keras' validation_split together with train_test_split().

```python
# model 1: single neural network

from tensorflow import keras
from keras import layers
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout, Activation, Flatten
from sklearn.model_selection import train_test_split

x_train = xc
y_train = yc

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.6, random_state=42)

model_single = Sequential()
model_single.add(Input(shape=(1,)))
model_single.add(Dense(10,activation = "relu"))
model_single.add(Dense(10,activation = "LeakyReLU"))
model_single.add(Dense(10,activation = "ThresholdedReLU"))
model_single.add(Dense(10,activation = "PReLU"))
model_single.add(Dense(10,activation = "ELU"))
model_single.add(Dense(10,activation = "relu"))
model_single.add(Dense(1))
# Compile the model
model_single.compile(loss = 'MAE', optimizer = 'Adam', metrics=['mae'])

# Train the model
model_single.fit(x_train, y_train, batch_size = 128, epochs = 50)

# Predictions on testing data
py = model_single.predict(x_test)

# Visualize predictions compared to actual values
plt.scatter(x_test, py)
plt.scatter(x_test, y_test)
plt.title("Model Fit")
plt.show()
es = []
for i in range(len(py)):
  es.append(y_test[i] - py[i])
plt.scatter(x_test, es)
plt.title("Error")
plt.show()

```

```python
# model 2: deeper neural network

model_deep = Sequential()
model_deep.add(Input(shape=(1,)))
model_deep.add(Dense(100,activation = "relu"))
model_deep.add(Dense(100,activation = "LeakyReLU"))
model_deep.add(Dense(100,activation = "ThresholdedReLU"))
model_deep.add(Dense(100,activation = "PReLU"))
model_deep.add(Dense(100,activation = "ELU"))
model_deep.add(Dense(100,activation = "relu"))
model_deep.add(Dense(1))

# Compile the model
model_deep.compile(loss = 'MAE', optimizer = 'Adam', metrics=['mae'])

# Train the model
model_deep.fit(x_train, y_train, batch_size = 128, epochs = 50)

# Predictions on testing data
py = model_deep.predict(x_test)

# Visualize predictions compared to actual values
plt.scatter(x_test, py)
plt.scatter(x_test, y_test)
plt.title("Model Fit")
plt.show()
es = []
for i in range(len(py)):
  es.append(y_test[i] - py[i])
plt.scatter(x_test, es)
plt.title("Error")
plt.show()
```

```python
# model 3: wider neural network

model_wide = Sequential()
model_wide.add(Input(shape=(1,)))
model_wide.add(Dense(1000,activation = "relu"))
model_wide.add(Dense(1000,activation = "LeakyReLU"))
model_wide.add(Dense(1000,activation = "ThresholdedReLU"))
model_wide.add(Dense(1000,activation = "PReLU"))
model_wide.add(Dense(1000,activation = "ELU"))
model_wide.add(Dense(1000,activation = "relu"))
model_wide.add(Dense(1))

# Compile the model
model_wide.compile(loss = 'MAE', optimizer = 'Adam', metrics=['mae'])

# Train the model
model_wide.fit(x_train, y_train, batch_size = 128, epochs = 50)

# Predictions on testing data
py = model_wide.predict(x_test)

# Visualize predictions compared to actual values
plt.scatter(x_test, py)
plt.scatter(x_test, y_test)
plt.title("Model Fit")
plt.show()
es = []
for i in range(len(py)):
  es.append(y_test[i] - py[i])
plt.scatter(x_test, es)
plt.title("Error")
plt.show()

```

#### Part 4 get model output and feedforward by yourself

Recall how a Feedforward Neural Network gets its output. Now choose your model with highest accuracy, and call get_weights( ) to get its weights and bias.
<br> Hint: bias and weights have different dimensions in most cases. Try to guess which index represent bias before you look it up on the internet.

<br> Afterwards, choose 5 data from your training dataset, do all the necessary calculations (with program of course), and get the output of your model. Compare it to the model.predict( ) result.
<br> They should be identical for the first several digitals, if not for all digits.

```python


# Get the weights and biases of the chosen model
weights = model_single.get_weights()

# Separate weights and biases
weights_list = weights[::2]
biases_list = weights[1::2]

# Choose 5 data points from the training dataset
sample_indices = [0, 100, 200, 300, 399]
sample_data = x_train[sample_indices]

# Perform feedforward calculations manually
outputs_manually_calculated = []
for data in sample_data:
    output = data
    for layer_weights, layer_biases in zip(weights_list, biases_list):
        output = np.dot(output, layer_weights) + layer_biases
        output = np.maximum(0, output)  # ReLU activation function for hidden layers
    outputs_manually_calculated.append(output)

# Get the output predictions using model.predict()
outputs_from_model = model_single.predict(sample_data)

# Compare the outputs
for i in range(len(sample_indices)):
    print("Sample", i+1)
    print("Manually Calculated Output:", outputs_manually_calculated[i])
    print("Output from model.predict():", outputs_from_model[i])
    print()


```
